```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the data from the CSV file
final_df = pd.read_csv('nba_final.csv')

# Filter the DataFrame to include only the specified columns
columns = ['H_W_PCT', 'H_FG_PCT', 'H_REB', 'H_AST', 'H_TOV', 'H_STL', 'H_BLK', 'H_PLUS_MINUS','H_OFF_RATING', 'H_DEF_RATING', 'H_TS_PCT', 'eloh_pre','Result']
# Calculate the correlation matrix
correlation = final_df[columns].corr()

# Plot the heatmap
plt.figure(figsize=(25, 25))
sns.heatmap(correlation, annot=True, fmt=".2f", cmap="coolwarm", square=True, cbar_kws={"shrink": .82})
plt.title('Correlation Matrix Heatmap')
plt.show()

# Display the correlation matrix
correlation
```
